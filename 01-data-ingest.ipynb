```python
from pathlib import Path
import pandas as pd
import json

# --- Paths ---
DATA_DIR = Path("../data/archive")  # Kaggle dataset unzipped here
MANIFEST_FILE = Path("../data/dataset_manifest.json")

# --- List CSV files ---
csv_files = list(DATA_DIR.glob("*.csv"))
print("CSV files found in archive:")
for f in csv_files:
    print("-", f.name)

# --- Quick preview of first file ---
if csv_files:
    sample_file = csv_files[0]
    df_sample = pd.read_csv(sample_file, nrows=5)
    print("\nSample data from:", sample_file.name)
    display(df_sample.head())
else:
    print("No CSV files found in", DATA_DIR)

# --- Save manifest for later use ---
manifest = {
    "dataset": "Crime in India (Kaggle - rajanand)",
    "files": [f.name for f in csv_files],
    "source": "https://www.kaggle.com/datasets/rajanand/crime-in-india"
}

with open(MANIFEST_FILE, "w") as f:
    json.dump(manifest, f, indent=4)

print("\nManifest saved at:", MANIFEST_FILE)
```
