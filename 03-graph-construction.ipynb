```python
# 03 - Graph construction, GNN training & evaluation
# Run in the same environment that has torch and torch_geometric installed.

import pandas as pd
import numpy as np
from pathlib import Path
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GATConv
import os

# Paths
DATA_DIR = Path("../data")
MODELS_DIR = Path("../models")
MODELS_DIR.mkdir(exist_ok=True)
CLEAN_FILE = DATA_DIR / "CrimeNet_master.csv"

# ------------- 1) Load cleaned dataset -------------
df = pd.read_csv(CLEAN_FILE)
print("Loaded:", df.shape)
display(df.head(3))

# ------------- 2) Feature engineering (district-level summary) -------------
# We'll aggregate by district because population & state info is missing in many rows.
agg_funcs = {
    "murder": ["mean", "std", "last"],
    "rape": ["mean", "std", "last"],
    "robbery": ["mean", "std", "last"],
    "dacoity": ["mean", "std", "last"],
    "arson": ["mean", "std", "last"],
    "theft": ["mean", "std", "last"],
    "other_ipc_crimes": ["mean", "std", "last"],
    "total_ipc_crimes": ["mean", "std", "last"]
}

# helper to compute last value per district (max year)
lasts = df.sort_values("year").groupby("district").last().reset_index()[["district","total_ipc_crimes"]].rename(columns={"total_ipc_crimes":"total_last"})
group = df.groupby("district")

features = []
for district, g in group:
    row = {"district": district}
    for col in ["murder","rape","robbery","dacoity","arson","theft","other_ipc_crimes","total_ipc_crimes"]:
        col_vals = g[col].fillna(0).values
        row[f"{col}_mean"] = col_vals.mean()
        row[f"{col}_std"]  = col_vals.std(ddof=0) if len(col_vals)>0 else 0.0
        row[f"{col}_last"] = col_vals[-1] if len(col_vals)>0 else 0.0
    # simple trend: slope using linear regression on years vs total_ipc_crimes
    try:
        years = g["year"].astype(float).values
        totals = g["total_ipc_crimes"].astype(float).values
        if len(years) > 1:
            coef = np.polyfit(years, totals, 1)[0]  # slope
        else:
            coef = 0.0
    except Exception:
        coef = 0.0
    row["total_trend"] = coef
    features.append(row)

feat_df = pd.DataFrame(features).reset_index(drop=True)
print("District-level features shape:", feat_df.shape)
display(feat_df.head())

# ------------- 3) Create hotspot label -------------
# We'll use average total_ipc_crimes mean per district to define hotspots (top 25%).
feat_df["label_score"] = feat_df["total_ipc_crimes_mean"]
threshold = feat_df["label_score"].quantile(0.75)
feat_df["is_hotspot"] = (feat_df["label_score"] >= threshold).astype(int)
print("Hotspot threshold (75th percentile):", threshold)
print("Hotspot distribution:\n", feat_df["is_hotspot"].value_counts(normalize=False))

# ------------- 4) Prepare feature matrix -------------
feature_cols = [c for c in feat_df.columns if c.endswith("_mean") or c.endswith("_std") or c.endswith("_last") or c=="total_trend"]
X = feat_df[feature_cols].fillna(0).values
y = feat_df["is_hotspot"].values
district_names = feat_df["district"].tolist()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ------------- 5) Build edges via k-NN on features (cosine or euclidean) -------------
k = 8  # neighbors per node
nbrs = NearestNeighbors(n_neighbors=k+1, metric="cosine").fit(X_scaled)  # +1 because neighbor includes itself
distances, indices = nbrs.kneighbors(X_scaled)

edge_index = []
num_nodes = X_scaled.shape[0]
for i in range(num_nodes):
    for j_idx in indices[i][1:]:  # skip the first neighbor (self)
        edge_index.append([i, int(j_idx)])
        edge_index.append([int(j_idx), i])  # add both directions

edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
print("Edge index shape:", edge_index.shape, "| Num edges (directed):", edge_index.shape[1])

# ------------- 6) Build PyG Data object -------------
x = torch.tensor(X_scaled, dtype=torch.float)
y_t = torch.tensor(y, dtype=torch.long)

data = Data(x=x, edge_index=edge_index, y=y_t)
print(data)

# ------------- 7) Train/Val/Test split (stratified) -------------
# We'll do node-level split (stratified by hotspot label)
idx = np.arange(num_nodes)
train_idx, test_idx = train_test_split(idx, test_size=0.2, stratify=y, random_state=42)
train_idx, val_idx = train_test_split(train_idx, test_size=0.2, stratify=y[train_idx], random_state=42)

train_mask = torch.zeros(num_nodes, dtype=torch.bool)
val_mask   = torch.zeros(num_nodes, dtype=torch.bool)
test_mask  = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[train_idx] = True
val_mask[val_idx]     = True
test_mask[test_idx]   = True

data.train_mask = train_mask
data.val_mask = val_mask
data.test_mask = test_mask

print("Train/Val/Test sizes:", train_mask.sum().item(), val_mask.sum().item(), test_mask.sum().item())

# ------------- 8) Define GAT model -------------
class GATNet(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, heads=4, dropout=0.3):
        super().__init__()
        self.conv1 = GATConv(in_channels, hidden_channels, heads=heads, concat=True)
        self.conv2 = GATConv(hidden_channels * heads, out_channels, heads=1, concat=False)
        self.dropout = dropout

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.elu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.conv2(x, edge_index)
        return x

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GATNet(in_channels=x.shape[1], hidden_channels=32, out_channels=2, heads=4, dropout=0.3).to(device)
data = data.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

# ------------- 9) Training loop with early stopping -------------
best_val_loss = float('inf')
best_model_state = None
patience = 30
patience_counter = 0
num_epochs = 400

for epoch in range(1, num_epochs+1):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()

    # validation
    model.eval()
    with torch.no_grad():
        val_out = model(data.x, data.edge_index)
        val_loss = criterion(val_out[data.val_mask], data.y[data.val_mask]).item()
    if val_loss < best_val_loss - 1e-6:
        best_val_loss = val_loss
        best_model_state = {k: v.cpu() for k, v in model.state_dict().items()}
        patience_counter = 0
    else:
        patience_counter += 1

    if epoch % 20 == 0 or epoch == 1:
        train_acc = accuracy_score(data.y[data.train_mask].cpu(), out[data.train_mask].argmax(dim=1).cpu())
        val_acc = accuracy_score(data.y[data.val_mask].cpu(), val_out[data.val_mask].argmax(dim=1).cpu())
        print(f"Epoch {epoch:03d} | Train loss: {loss.item():.4f} | Val loss: {val_loss:.4f} | Train acc: {train_acc:.4f} | Val acc: {val_acc:.4f}")

    if patience_counter >= patience:
        print("Early stopping at epoch", epoch)
        break

# Load best model state
model.load_state_dict(best_model_state)
torch.save(model.state_dict(), MODELS_DIR / "best_gat_district.pt")
print("Saved best model to:", MODELS_DIR / "best_gat_district.pt")

# ------------- 10) Final Evaluation on Test set -------------
model.eval()
with torch.no_grad():
    logits = model(data.x, data.edge_index)
    probs = F.softmax(logits, dim=1)[:,1].cpu().numpy()  # probability of class 1
    preds = logits.argmax(dim=1).cpu().numpy()

y_true = data.y.cpu().numpy()
y_test = y_true[test_mask.cpu().numpy()]
y_pred = preds[test_mask.cpu().numpy()]
y_score = probs[test_mask.cpu().numpy()]

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
auc = roc_auc_score(y_test, y_score) if len(np.unique(y_test)) > 1 else float("nan")
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel() if cm.size == 4 else (0,0,0,0)

print("\n--- Test Metrics ---")
print(f"Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f} | AUC: {auc:.4f}")
print(f"TP={tp}, TN={tn}, FP={fp}, FN={fn}")
print("Confusion matrix:\n", cm)

# Plot ROC
if not np.isnan(auc):
    fpr, tpr, _ = roc_curve(y_test, y_score)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve (Test)")
    plt.legend()
    plt.show()

# Save predictions for inspection
results_df = pd.DataFrame({
    "district": district_names,
    "label": y_true,
    "pred": preds,
    "prob": probs
})
results_df.to_csv(DATA_DIR / "gnn_district_predictions.csv", index=False)
print("Saved predictions to:", DATA_DIR / "gnn_district_predictions.csv")
```

```python
# ===== Baseline comparison: RandomForest & XGBoost =====
import joblib
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, roc_curve, confusion_matrix)
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns

MODELS_DIR = Path("../models")
MODELS_DIR.mkdir(exist_ok=True)

# We assume these objects exist from previous cell:
# feat_df (district-level dataframe)
# feature_cols (columns used), X_scaled (numpy array), y (labels)
# train_idx, val_idx, test_idx (numpy arrays of indices)
# and GAT model predictions: preds (all nodes), probs (prob for class 1)

# If preds/probs are torch/tensors, convert to numpy
try:
    gat_preds_all = preds
    gat_probs_all = probs
except NameError:
    # if not present, compute them using saved predictions file
    res = pd.read_csv("../data/gnn_district_predictions.csv")
    gat_preds_all = res["pred"].values
    gat_probs_all = res["prob"].values

# Train/val/test splits
X_train = X_scaled[train_idx]
y_train = y[train_idx]
X_val   = X_scaled[val_idx]
y_val   = y[val_idx]
X_test  = X_scaled[test_idx]
y_test  = y[test_idx]

results = []

def evaluate_model(name, y_true, y_pred, y_score):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    auc = roc_auc_score(y_true, y_score) if len(np.unique(y_true))>1 else np.nan
    cm = confusion_matrix(y_true, y_pred)
    if cm.size == 4:
        tn, fp, fn, tp = cm.ravel()
    else:
        tn = fp = fn = tp = 0
    return {"model": name, "accuracy": acc, "precision": prec, "recall": rec, "f1": f1, "auc": auc,
            "tp": int(tp), "tn": int(tn), "fp": int(fp), "fn": int(fn)}

# --- 1) RandomForest ---
rf = RandomForestClassifier(n_estimators=300, max_depth=12, random_state=42, n_jobs=-1, class_weight='balanced')
rf.fit(X_train, y_train)
rf_probs_test = rf.predict_proba(X_test)[:,1]
rf_preds_test = (rf_probs_test >= 0.5).astype(int)

res_rf = evaluate_model("RandomForest", y_test, rf_preds_test, rf_probs_test)
results.append(res_rf)
joblib.dump(rf, MODELS_DIR / "rf_district.pkl")
print("Saved RandomForest to", MODELS_DIR / "rf_district.pkl")

# --- 2) XGBoost (version-safe) ---
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    use_label_encoder=False,
    eval_metric="logloss",
    random_state=42
)

# Fit without early stopping for compatibility
xgb.fit(X_train, y_train)

xgb_probs_test = xgb.predict_proba(X_test)[:, 1]
xgb_preds_test = (xgb_probs_test >= 0.5).astype(int)

res_xgb = evaluate_model("XGBoost", y_test, xgb_preds_test, xgb_probs_test)
results.append(res_xgb)
joblib.dump(xgb, MODELS_DIR / "xgb_district.pkl")
print("Saved XGBoost to", MODELS_DIR / "xgb_district.pkl")


# --- 3) GAT (use earlier preds/probs) ---
# We only evaluate on test indices
gat_preds_test = gat_preds_all[test_idx]
gat_probs_test = gat_probs_all[test_idx]
res_gat = evaluate_model("GAT", y_test, gat_preds_test, gat_probs_test)
results.append(res_gat)

# --- Compile results table ---
res_df = pd.DataFrame(results).set_index("model")
print("\n=== Model comparison on TEST set ===")
display(res_df[["accuracy","precision","recall","f1","auc","tp","tn","fp","fn"]])

# Save comparison
res_df.to_csv(MODELS_DIR / "model_comparison.csv")
print("Saved model comparison to:", MODELS_DIR / "model_comparison.csv")

# --- Plot ROC curves together ---
plt.figure(figsize=(7,6))
for name, probs, preds in [("RandomForest", rf_probs_test, rf_preds_test), ("XGBoost", xgb_probs_test, xgb_preds_test), ("GAT", gat_probs_test, gat_preds_test)]:
    if len(np.unique(y_test))>1:
        fpr, tpr, _ = roc_curve(y_test, probs)
        auc_val = roc_auc_score(y_test, probs)
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc_val:.3f})")
plt.plot([0,1],[0,1],'k--', alpha=0.4)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Test Set")
plt.legend()
plt.grid(True)
plt.show()

# --- Feature importance (RandomForest) ---
fi = pd.Series(rf.feature_importances_, index=feature_cols).sort_values(ascending=False)[:20]
plt.figure(figsize=(8,6))
sns.barplot(x=fi.values, y=fi.index)
plt.title("Top 20 Feature Importances (RandomForest)")
plt.tight_layout()
plt.show()

# Print quick recommendation based on results
best = res_df.sort_values("f1", ascending=False).index[0]
print(f"\nRecommendation: Best model by F1 = {best}. If GAT remains best, keep graph pipeline; otherwise investigate features/edges/hyperparams further.")
```
